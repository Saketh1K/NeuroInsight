```python
import tensorflow as tf

# Paths to your saved models
resnet_model_path = "D://NeuroCure-main//NeuroCure-main//notebooks//ResNet50V2.keras"
custom_model_path = "D://NeuroCure-main//NeuroCure-main//notebooks//custom_model.keras"

segmentation_model_path = "D://NeuroCure-main//NeuroCure-Main//notebooks//best_model.keras"


# Load models
resnet_model = tf.keras.models.load_model(resnet_model_path)
custom_model = tf.keras.models.load_model(custom_model_path)


# For segmentation model, pass the custom_objects parameter

def dice_loss(y_true, y_pred, smooth=1e-6, weight_bg=0.2, weight_fg=0.8):
        y_true_f = tf.keras.backend.flatten(y_true)
        y_pred_f = tf.keras.backend.flatten(y_pred)

        # Compute the intersection
        intersection = tf.keras.backend.sum(y_true_f * y_pred_f)

        # Compute the Dice coefficient
        dice = (2. * intersection + smooth) / (tf.keras.backend.sum(y_true_f) + 
                                               tf.keras.backend.sum(y_pred_f) + smooth)

        return 1 - dice

segmentation_model = tf.keras.models.load_model(segmentation_model_path, custom_objects={'dice_loss': dice_loss})

# Print architectures using .summary()
print("ResNet50V2 Model Architecture:")
resnet_model.summary()

print("\nCustom Model  Architecture:")
custom_model.summary()




print("\nSegmentation Model Architecture:")
segmentation_model.summary()
```
